Lagrange polynomial interpolation finds a single polynomial that goes through all the data points.

Below we will show two examples of the Lagrange polynomial interpolation using Python. 

Example 1: Find the Lagrange basis polynomials for the data set x = [1, 2, 3] and y = [2, 5, 4]. Plot each polynomial and verify that Pi(xj) = 1 when i = j and Pi(xj) = 0 when i =/ j. 


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
```

```python
x = [1, 2, 3]
y = [2, 5, 4]

x_new = np.arange(0, 4.2, .1)

f = lagrange(x, y)
```

```python
fig = plt.figure(figsize = (10,8))
plt.plot(x_new, f(x_new), 'b', x, y, 'ro')
plt.title('Lagrange Polynomial Interpolation')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Example 2: Find the lagrange basis polynomials for the data set x = [4, 5, 6] and y = [7, 9, 8]. Plot each polynomial and verify the property that Pi(xj) = 1 when i = j ans Pi(xj) = 0 when i =/ j. 

```python
x_two = [4, 5, 6]
y_two = [5, 7, 6]

x_new = np.arange(1, 9.6, .5)

f_two = lagrange(x_two, y_two)
```

```python
plt.plot(x_new, f_two(x_new), 'b', x_two, y_two, 'ro')
plt.title('Lagrange Polynomial Interpolation')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Newton's Polynomial Interpolation is another popular method to fit exactly for a set of data points. 

Below we will show an example of Newton's polynomial interpolation using Python.

Example 3: Calculate the divided differences table for x = [-6, -2, 0, 3] and y = [-3, 5, 0, 2].

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

%matplotlib inline
```

```python
def divided_diff(x, y):

  n = len(y)
  coef = np.zeros([n, n])
  coef[:,0] = y

  for j in range (1,n):
    for i in range(n-j):
      coef[i][j] = \
      (coef [i+1][j-1] - coef[i][j-1]) / (x[i+j]-x[i])

  return coef

def newton_poly(coef, x_data, x):

  n = len(x_data) - 1
  p = coef[n]
  for k in range (1, n+1):
    p = coef[n-k] + (x - x_data[n-k]) * p
  return p
```

```python
xthree = np.array([-6, -2, 0, 3])
ythree = np.array([-3, 5, 0, 2])

a_s = divided_diff(xthree, ythree) [0, :]

x_neww = np.arange(-6, 3.1, .1)
y_neww = newton_poly(a_s, xthree, x_neww)

plt.figure(figsize = (12,8))
plt.plot(xthree, ythree, 'bo')
plt.plot(x_neww, y_neww)
```
