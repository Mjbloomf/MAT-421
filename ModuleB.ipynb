Assume data of independent data values xi and dependent data vlaues yi, where i = 1, ..., n. The goal is to estimate a function that goes through the data points. No underlying model is required, but it must still inform the user quality.

To begin we will use python to creatw an example of interpolation.

Example 1: Find the interpolation at x = 2.5 on the data x = [1, 2, 3] and y = [3, 2, 0] using python.

```python
import numpy as np
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [3, 2, 0]

x_new = 2.5
y_new = np.interp(x_new, x, y)
print(y_new)
```

```python
plt.plot(x, y, 'o-')
plt.title('Interpolation at x = 2.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```



In this section of homework we will also cover linear and cubic spline intepolation. 

Linear interpolation is the estimated point that is assumed to lie on the line joining the nearest point to the left and right. 

Below we will show an example of linear interpolation. 

Example 2: Find the linear interpolation at x = 2.5 based on data x = [1, 2, 3], y = [2, 6, 8]. Create a graph.

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
```

```python
x = [1, 2, 3]
y = [2, 6, 8]

f = interp1d(x, y)
y_hat = f(2.5)
print(y_hat)
```

```python
plt.figure(figsize = (10, 8))
plt.plot(x, y, '-ob')
plt.plot(2.5, y_hat, 'ro')
plt.title('Linear Interpolation at x = 2.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Now we will discuss cubic spline interpolation.

Cubic spline interpolation is the interpolating function as a set of piecewise cubic funtions. 

Example 2: Use the CubicSpline function to plot the cubic spline interpolation of the data set x = [1,2,3] and y = [4,6,5] for 0 <= x <= 4. 



```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
y = [4, 6, 5]

f = CubicSpline(x, y, bc_type = 'natural')
x_new = np.linspace(0, 4, 100)
y_new = f(x_new)
```

```python
plt.figure(figsize = (10, 8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
