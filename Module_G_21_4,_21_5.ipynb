```
import numpy as np

a = 0

b = np.pi

n = 14

h = (b - a) / (n - 1)

x1 = np.linspace(a, b, n)

f = np.cos(x1)



Isim = (h/3) * (f[0] + 2*sum(f[:n-2:2]) \
            + 4*sum(f[1:n-1:2]) + f[n-1])
errsim = 2 - Isim

print(Isim)
print(errsim)
```

```
from scipy.integrate import cumtrapz
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('seaborn-poster')

x = np.arange(.1, np.pi, 0.01)

Fex = np.sin(x)

Fapp = cumtrapz(np.cos(x), x)


plt.figure(figsize = (14,10))
plt.plot(x, Fex)
plt.plot(x[1::], Fapp)
plt.grid()
plt.tight_layout()
plt.title('$F(x) = \int_0^{x} cos(y) dy$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['Exact', 'Approximation'])
plt.show()
```
